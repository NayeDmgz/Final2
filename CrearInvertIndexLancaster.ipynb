```python
# El programa lee las url de un txt y entramos a esa pagina 
#para poder recuperar las palabras y contar su frecuencia
#posteriomente el diccionario se guarda en un txt llamado raiz_ind_inv.
# Nayeli Itzel Dominguez Avila

# librerias necesarias
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
#Para tomar urls sin certificado ssl
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

#definimos los contenedores del diccionario y una lista de url
diccionario = {}
url = []

#leemos las urls del txt
with open("urls.txt",'r',encoding='utf-8') as archivo:
    for lines in archivo:
        #guardamos las url en una lista de strings
        url.append(lines) 

#Recorremos las url de la lista url
for i in url:
    frecuencia_palabras = []
    results = []
    #impresion para observar las url leidas
    print(i)
    html = urlopen(i).read()
    soup = BeautifulSoup(html,features="lxml")
    #eliminamos estilos y guiones
    for script in soup(["script", "style"]):
        script.extract()    # extraccion

    #obtenemos el texto del url
    text = soup.get_text()

    #Dividimos cada linea y quitamos los espacios del inicio y el final
    lines = (line.strip() for line in text.splitlines())
    #Dividimos cada pedazo de texto
    chunks = (phrase.strip() for line in lines for phrase in line.split('  '))
    #Evitamos las lineas en blanco
    text = ' '.join(chunk for chunk in chunks if chunk)
    # separar por palabras
    texto = text.split(' ') 
    #For para recorrer las palabras encontradas
    for w in texto: 
        frecuencia_palabras.append([w,texto.count(w)])
    #Contamos la frecuencai de aparicion de las palbras y definimos una lista vacia
    for w in frecuencia_palabras:
        w[0] = re.sub(r'[^\w\s]','',w[0])
        w[0] = w[0].lower()
    frecuencia_palabras2 = []
    #Vamos a単adiendo las frecuencais con cada palabra
    for w in frecuencia_palabras:
        frecuencia_palabras2.append((w[0],w[1]))
     #Se elimilarn palabras repetidas pero sin borrar la cantidad de palabras
    for item in frecuencia_palabras2:
        if item not in results:
            if len(item[0])>2:
                if len(item[0])<20:
                    if not item[0].isdigit():
                        results.append(item)
    #Guardamos en el diccionario
    diccionario[i] = results 

#Invertir diccionario
new_diccionario = {} #Diccionario Invertido
lista = []
#Recorremos el diciconario y lo a単adimos a la lista nueva
for item in diccionario:
    for palab in diccionario[item]:
            lista.append([palab[0],item, palab[1]])
#Definimos los indices y se a単aden
indi = len(lista)
palabras = []
for i in range(indi):
    if lista[i][0] not in palabras:
        palabras.append([lista[i][0]])
#Organizamos en invertido
for i in palabras:
    for j in lista:
        if j[0] in i:
            if j[1] not in i:
                i.append(j[1])
                i.append(j[2])
#definimos una lista nueva para guardar los resultados
results = []
for item in palabras:
    if item not in results:
        results.append(item)
#Guardamos en el diccionario y ordenamos por su tama単o
for i in results:
    if len(i) == 3:
        new_diccionario[i[0]] = (i[1],i[2])
    elif len(i) == 5:
        new_diccionario[i[0]] = (i[1],i[2],i[3],i[4])
    elif len(i) == 7:
        new_diccionario[i[0]] = (i[1],i[2],i[3],i[4],i[5],i[6])

#Guardamos en un txt el diccionario
with open("raiz_ind_inv.txt", 'w', encoding='utf-8') as archivo: 
    archivo.write(str(new_diccionario))
```
